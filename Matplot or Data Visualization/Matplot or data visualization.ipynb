# Bar Chart

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[6,7,8,9,10]
plt.xlabel("height",fontsize=8)
plt.ylabel("digit",fontsize=8)
plt.title("chart",fontsize=8)
C=["y","g","r","pink","b"]    #Color names...black color name=k
plt.bar(x,y,width=0.5,color=C,align="center",edgecolor="black")
plt.show()
```

# Stacked Plot

```python
import matplotlib.pyplot as plt
import pandas as pd
x=[2,3,4,5]
y=[7,8,9,7]

df=pd.DataFrame({"x":[2,3,4,5],"y":[7,8,9,7]})
ax=df.plot.bar(stacked=True)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
age=[11,12,13,14,15,16,17,18,19]
height=[25,26,27,28,29,30,31,32,33]

plt.scatter(age,height,c=height,s=50,edgecolor="k",cmap="Greens")

cbar=plt.colorbar()
cbar.set_label("height")
```

# Line Chart

```python
import matplotlib.pyplot as plt
age=[11,12,13,14,15,16,17,18,19]
height=[25,26,27,28,29,30,31,32,33]
weight=[23,25,28,30,26,27,26,25,24]
plt.plot(age,height,color="Black",marker="*",linestyle="--")   #marker = o&* are the denote the points
plt.plot(age,weight,color="Green",marker="o")
plt.legend(["Height","Weight"])
```

# Pie Chart 

```python
student=[60,20,30,10,10]
Class=["Python","c","Java","ML","C++"]
plt.pie(student,labels=Class,autopct="%1.2f%%")
plt.title("Pie Chart")
plt.tight_layout()
```

```python
import matplotlib.pyplot as plt
lab='Frogs','Hogs','Dogs','Logs'
size=[15,30,45,10]
ex=(0.2,0,0.1,0.)
fig1,a1=plt.subplots()

a1.pie(size,explode=ex,labels=lab,autopct='%1.1f%%',shadow=True,startangle=90)
plt.show()
```
